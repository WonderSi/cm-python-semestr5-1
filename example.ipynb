1a) 
Определить предельную абсолютную погрешность по известным точным x и приближенным x* числам

```python
# x = π/15
# x* = 0,20943

# Δ(x*) = | x - x* |

x_true_1 = math.pi / 15
x_approx_1 = 0.20943
delta_1a = abs(x_true_1 - x_approx_1)
print(f"Предельная абсолютная погрешность: {delta_1a:.5f}")
```

1b) Определить предельную 
относительную погрешность по известным точным x, приближенным x* числам или абсолютной погрешности Δ(x*)

```python
# x*=13,163845
# Δ(x*) = 2,5 * 10^(-4); 0,00025

# δ(x*) = | x - x* | / | x* |
# δ(x*) = Δ(x*) / | x* |

x_approx_1b = 13.163845
delta_1b = 2.5e-4
rel_error_1b = delta_1b / abs(x_approx_1b)
print(f"Предельная относительная погрешность: {rel_error_1b:.2e} (или {rel_error_1b * 100:.4f}%)")
```

2 ) Определить верные и значащие цифры числа по известному приближенному числу x* и его абсолютной погрешности Δ(x*)

```python
# x* = 0,0123691500 
# Δ(x*) = 10^-9;    0,000000001

# Сравним:
# 0,0123691500
# 0,000000001

# Δ(x*) ≤ 0.5 * 10^m
# Единица разряда:      0,000000001 
# Половина:             0,0000000005
# delta(x*) должно быть меньше половины текущего разряда
# 0,000000001 > 0,0000000005 => Не включаем в состав верных цифр

def analyze_significant_digits(x_input, delta_val):
    """
    Определеяет верные значащие цифр после числа X с погрешностью Δ(x*)
    по строгому критерию (Δ(x*) ≤ 0.5 * 10^m)
    """
    print(f"Анализ числа: ", {x_input})
    print(f"Абсолютная погрешность: ", {delta_val})

    raw_digits_str = x_input.replace('.', '').lstrip('0')
    all_significant_digits = list(raw_digits_str)

    x_val = float(x_input)

    if x_val == 0:
        print("Анализ невозможен, число равно 0")
        return

    # Находим разряд первой значащей цифры (слева)
    # m_start = [log(10)|x|]
    # math.floor() - Округление вниз
    m_start = math.floor(math.log10(abs(x_val)))

    # Находим разряд последней значащей цифры (справа)
    # Используем формулу: Δ(x*) ≤ 0.5 * 10^m
    # преобразуем формулу, чтобы найти m:
    # ↓ 2 * Δ(x*) ≤ 10^m
    # ↓ [log(10)2 * Δ(x*)] ≤ m
    # math.ceil() - Округление вверх
    m_end = math.ceil(math.log10(2 * delta_val))
    
    # Считаем количесвто верных цифр
    # Пример:
    # Если m_start = -2, m_end = -8, то их (-2) - (-8) + 1 = 7
    count_valid = m_start - m_end + 1

    if count_valid <= 0:
        print("В этом числе нет верных цифр (погрешность слишком велика)!")
        return
    
    # Формируем строку для вывода самих цифр
    # Формтируем число как строку с нужной точностью
    decimal_places = max(0, -m_end)
    formatted_x = f"{x_val:.{decimal_places}}"

    # Мы получили строку и нам нужно почистить строку, 
    # чтобы получить только список значащих цифр
    # Убираем точку и ведущие нули 
    clean_digits_str = formatted_x.replace('.', '').lstrip('0')

    # [Для надежности]
    # Берем нужное количество цифр
    final_digits = list(clean_digits_str[:count_valid])

    print("-" * 20)
    print(f"Разряд пeрвой цифры: 10^{m_start}")
    print(f"Разряд последней цифры: 10^{m_end}")
    print(f"Значащие цифры: {', '.join(final_digits)}")
    print(f"Верные цифры: {', '.join(all_significant_digits)}")
    print("-" * 20)

    return
    pass



x = "0.0123691500"
delta = 1e-9

analyze_significant_digits(x, delta)
```
